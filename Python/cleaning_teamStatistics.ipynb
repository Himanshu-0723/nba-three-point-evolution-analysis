```python
import pandas as pd
```

```python
team_stats = pd.read_csv(r"C:\Users\himan\Downloads\DA\Project\NBA_Analytics\Data\Raw\TeamStatistics.csv")
```

```python
team_stats.columns
```

```python
team_stats.shape
```

```python
team_stats['gameDateTimeEst'] = pd.to_datetime(
    team_stats['gameDateTimeEst'],
    format = 'ISO8601',
    utc = True    
)
```

```python
team_stats['gameDateTime'] = pd.to_datetime(team_stats['gameDateTimeEst']).dt.tz_localize(None)
```

```python
team_stats['season'] = team_stats['gameDateTime'].apply(
    lambda x: x.year if x.month >= 10 else x.year-1
)
```

```python
team_stats = team_stats[team_stats['season'] >= 2001]
```

```python
team_stats['team'] = team_stats['teamCity'] + " " + team_stats['teamName']
```

```python
team_stats['opponentTeam'] = team_stats['opponentTeamCity'] + " " +team_stats['opponentTeamName']
```

```python
def convert_minutes(x):
    if pd.isna(x):
        return 0
    if isinstance(x, str) and ':' in x:
        m, s = x.split(':')
        return int(m) + int(s)/60
    return float(x)

team_stats['numMinutes'] = team_stats['numMinutes'].apply(convert_minutes)
```

```python
team_stats['FG_pct'] = (
    team_stats['fieldGoalsMade']/
    team_stats['fieldGoalsAttempted']
) 

team_stats['FT_pct'] = (
    team_stats['freeThrowsMade']/
    team_stats['freeThrowsAttempted']
)

team_stats['3P_pct'] = (
    team_stats['threePointersMade']/
    team_stats['threePointersAttempted']
)
```

```python
team_stats = team_stats.drop(columns=[
    'gameDateTimeEst', 'teamCity', 'teamName', 'opponentTeamCity', 'opponentTeamName',
    'fieldGoalsPercentage', 'threePointersPercentage', 'freeThrowsPercentage'
])
```
