```python
import numpy as np
import pandas as pd
```

```python
players = pd.read_csv(r"C:\Users\himan\Downloads\DA\Project\NBA_Analytics\Data\Raw\Players.csv")
```

```python
players.columns
```

```python
players["name"] = players["firstName"] + " " + players["lastName"]
```

```python
def map_position(row):
    if row['guard']:
        return 'G'
    elif row['forward']:
        return 'F'
    elif row['center']:
        return 'C'
    else:
        return 'Unknown'

players['position'] = players.apply(map_position, axis=1)
```

```python
players = players.drop(columns=['firstName', 'lastName', 'birthdate', 'lastAttended', 
                                'country', 'height', 'bodyWeight', 'guard', 'forward', 'center', 
                                'draftYear', 'draftRound', 'draftNumber'])
```
